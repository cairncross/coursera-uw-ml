```python
import graphlab as gl
```

# Load the CIFAR-10 dataset

```python
image_train = gl.SFrame('image_train_data/')
```

```python
image_train.head()
```

# Train a nearest neighbour model for retrieving images using deep features

```python
knn_model = gl.nearest_neighbors.create(image_train,
                                       features=['deep_features'],
                                       label='id')
```

# Use image retrieval model with deep features to find similar images

```python
gl.canvas.set_target('ipynb')
cat = image_train[18:19]
cat['image'].show()
```

```python
knn_model.query(cat)
```

```python
def get_images_from_ids(query_result):
    return image_train.filter_by(query_result['reference_label'], 'id')
```

```python
cat_neighbours = get_images_from_ids(knn_model.query(cat))
```

```python
cat_neighbours['image'].show()
```

```python
car = image_train[8:9]
```

```python
car['image'].show()
```

```python
car_neighbours = get_images_from_ids(knn_model.query(car))
```

```python
car_neighbours['image'].show()
```

# Just for fun, let's create a lambda to find a show nearest neighbour images

```python
show_neighbours = lambda i: get_images_from_ids(knn_model.query(image_train[i:i+1]))['image'].show()
```

```python
show_neighbours(8)
```

```python
show_neighbours(28)
```

```python
show_neighbours(1000)
```
